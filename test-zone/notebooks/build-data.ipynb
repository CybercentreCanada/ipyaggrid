```python
# Consistent Dataframe for examples
import urllib.request, json 
with urllib.request.urlopen("https://data.nasa.gov/api/views/gh4g-9sfh/rows.json?accessType=DOWNLOAD") as url:
    meteorites = json.loads(url.read().decode())
```

```python
import requests as rq
i = 0
for meteorite in meteorites:
    i+=1
    if 'geolocation' in meteorite:
        lat = meteorite['latitude']
        lon = meteorite['longitude']
        url = "https://nominatim.openstreetmap.org/reverse?format=json&lat="+str(lat)+"&lon="+str(lon)+"&zoom=10&addressdetails=1"
        r = rq.get(url)
        json_res = json.loads(r.content.decode('utf-8'))
        if 'address' in json_res:
            try:
                meteorite['country'] = json_res['address']['country']
                meteorite['country_code'] = json_res['address']['country_code']
                print(c_to[meteorite['country_code']])
                meteorite['continent'] = continents[c_to[meteorite['country_code']]]
                res = 'Point('+str(meteorite['geolocation']['coordinates'][0])+','+str(meteorite['geolocation']['coordinates'][1])+')'
                meteorite['geolocation'] = res
            except:
                pass
    del meteorite['nametype']
    if i % 10 ==0:
        print(i)
```

```python
i=0
computed_meteorites = []
for meteorite in meteorites['data']:
    new_met = {}
    new_met['id'] = meteorite[0]
    new_met['name'] = meteorite[8]
    new_met['recclass'] = meteorite[11]
    new_met['mass'] = meteorite[12]
    new_met['year'] = meteorite[14]
    new_met['status'] = meteorite[13]
    new_met['latitude'] = meteorite[15]
    new_met['longitude'] = meteorite[16]
    lat = new_met['latitude']
    lon = new_met['longitude']
    url = "https://nominatim.openstreetmap.org/reverse?format=jsonv2&lat="+str(lat)+"&lon="+str(lon)+"&zoom=10&addressdetails=1"
    r = rq.get(url)
    try:
        json_res = json.loads(r.content.decode('utf-8'))
        if 'address' in json_res:
            try:
                new_met['country'] = json_res['address']['country']
                new_met['country_code'] = json_res['address']['country_code']
                new_met['continent'] = continents[c_to[new_met['country_code']]]
            except Exception as e:
                print(e)
    except:
        pass
    computed_meteorites.append(new_met)
    if i%100==0:
        print(i)
        print(new_met)
    i+=1
```
